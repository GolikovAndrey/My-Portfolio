```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import count
```

```python
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
df = spark.read.json('/content/world-cities.json')
```

```python
df.show()
```

## Задача:
- Вывести на экран схему кадра данных;
- Вычислять  и выводить на экран в порядке убывания количества городов страны сгруппированные по стране и региону;

Cхема датафрейма:

```python
df.printSchema()
df.createOrReplaceTempView('df')
```

Количество городов в странах:

```python
spark.sql('''
  select country, count(distinct subcountry) as subcountry, count(name) as ctn from df group by country order by ctn desc
''').show()
```
