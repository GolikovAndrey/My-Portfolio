## 4) Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
juns = titanic['Age'].loc[titanic['Age'] < 30]
older = titanic['Age'].loc[titanic['Age'] > 60]
```

```python
jsurv = juns.loc[titanic['Survived'] == 1]
osurv = older.loc[titanic['Survived'] == 1]
```

```python
jres = jsurv.count() * 100/juns.count()
ores = osurv.count() * 100/older.count()
```

```python
print(f"{round(jres, 1)}% среди молодых и {round(ores, 1)}% среди старых")
```

Правда

## 5) Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
men = titanic['Sex'].loc[titanic['Sex'] == 'male']
women = titanic['Sex'].loc[titanic['Sex'] == 'female']
```

```python
msurv = men.loc[titanic['Survived'] == 1]
fsurv = women.loc[titanic['Survived'] == 1]
```

Правда

## 8) Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

1.   В среднем мужчины 1-го класса старше 40 лет
2.   В среднем женщины 1-го класса старше 40 лет
3.   Мужчины всех классов в среднем старше женщин того же класса
4.   В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса

```python
m1c = titanic['Age'].loc[titanic['Sex'] == 'male'].loc[titanic['Pclass'] == 1].mean() # Мужчины 1-го класса
m1c
```

Верно

```python
w1c = titanic['Age'].loc[titanic['Sex'] == 'female'].loc[titanic['Pclass'] == 1].mean() # Женщины 1-го класса
w1c
```

Неверно

```python
m2c = titanic['Age'].loc[titanic['Sex'] == 'male'].loc[titanic['Pclass'] == 2].mean() # Мужчины 2-го класса
w2c = titanic['Age'].loc[titanic['Sex'] == 'female'].loc[titanic['Pclass'] == 2].mean() # Женщины 2-го класса
m3c = titanic['Age'].loc[titanic['Sex'] == 'male'].loc[titanic['Pclass'] == 3].mean() # Мужчины 3-го класса
w3c = titanic['Age'].loc[titanic['Sex'] == 'female'].loc[titanic['Pclass'] == 3].mean() # Женщины 3-го класса

m1c > w1c, m2c > w2c, m3c > w3c
```

Верно

```python
firstClass = titanic['Age'].loc[titanic['Pclass'] == 1].mean()
secondClass = titanic['Age'].loc[titanic['Pclass'] == 2].mean()
thirdClass = titanic['Age'].loc[titanic['Pclass'] == 3].mean()

firstClass > secondClass > thirdClass
```

Верно
