```python
def parseLine(line):
    return line.split(',')
```

```python
parseIndexedDoubleCol = lambda list: [float(value) for value in list]
```

```python
def average(list):
    return sum(list)/len(list)
```

```python
with open('/content/california_housing_test.csv') as file:
    columns = [2, 3, 5]

    file.readline()
    list = [parseIndexedDoubleCol(parseLine(line)) for line in file]

    list1 = [] # забыл, как это можно сделать генератором списка
    list2 = [] # забыл, как это можно сделать генератором списка
    list3 = [] # забыл, как это можно сделать генератором списка
    
    for i in range(len(list)):
        for j in range(len(list[i])):
            if j == columns[0]:
                list1.append(list[i][j])
            if j == columns[1]:
                list2.append(list[i][j])
            if j == columns[2]:
                list3.append(list[i][j])

    print(round(average(list1), 2), round(average(list2), 2), round(average(list3), 2))
```
