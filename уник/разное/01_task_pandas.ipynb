```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/aksenov7/Kaggle_competition_group/master/adult.data.csv")
data.head()
```

```python
# def married(row):
#   return "Married" in row
data["married"] = data["marital-status"].apply(lambda row: "Married" in row)
```

```python
data
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
count = len([male for male in data['sex'] if male == 'Male'])
print(f'Men: {count}, women: {len(data) - count}')
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
print(f"Mean women age: {data['age'].where(data['sex'] == 'Female').mean()}")
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
share = data[data['native-country'] == 'Germany'].count()[0]
all = data.count()[0]
print(f'Share of Germany citizens: {share/all}')
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
print(f"Greater than 50k mean: {data['age'].loc[data['salary'] == '>50K'].describe()[1]}, std: {data['age'].loc[data['salary'] == '>50K'].describe()[2]}")
print(f"Less or equal than 50k mean: {data['age'].loc[data['salary'] == '<=50K'].describe()[1]}, std: {data['age'].loc[data['salary'] == '<=50K'].describe()[2]}")
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
data[data['education'].isin(['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate'])].count()[0] == data[data['salary'] == '>50K'].count()[0]
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
print(f"Describe: {data.groupby(['race', 'sex'])['age'].describe()}")
print(f"Max age: {data['age'][(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')].max()}")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
richMen = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]
married = richMen[richMen['marital-status'].isin(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'])].count()[0]
unmarried = richMen.count()[0] - married
print(f'Married rich men: {married}, unmarried rich men: {unmarried}')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max = data['hours-per-week'].max()
same = data[data['hours-per-week'] == max]
same_but_rich = same[same['salary'] == '>50K']

print(f"Max hours per week: {max}, people with same hours: {same.count()[0]}, same hours and reach people: {same_but_rich.count()[0]/same.count()[0]*100}%")
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
less50 = data[data['salary'] == '<=50K']
greater50 = data[data['salary'] == '>50K']

less50['hours-per-week'].groupby(less50['native-country']).mean()
```

```python
greater50['hours-per-week'].groupby(greater50['native-country']).mean()
```
